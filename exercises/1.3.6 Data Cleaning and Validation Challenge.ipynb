```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
%matplotlib inline
```

```python
journals = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding = "ISO-8859-1")
journals
```

```python
# Create separate series to clean data of journal titles
journal_title = journals.iloc[:, 2]
journal_title
```

```python
# This block of code is meant to take out all of the variations of 'journal' to 
# strip down to the the main words of the title
journal_title = journal_title.str.replace('journal of', '')
journal_title = journal_title.str.replace('journal', '')
journal_title = journal_title.str.replace('Journal of', '')
journal_title = journal_title.str.replace('J', '')
journal_title = journal_title.str.replace('Journal', '')
journal_title
```

```python
# strip white space
journal_title = journal_title.str.strip()
journal_title
```

```python
# make everything lowercase
journal_title = journal_title.str.lower()
journal_title
```

```python
# Insert cleaned data back into dataframe
journals.iloc[:, 2] = journal_title
journals
```

```python
# Group data by cleaned journal names
sorted_journals = journals.groupby('Journal title').count()
# Sort data by most articles per journal
sorted_journals.sort_values(by='Article title', ascending=False)
```

```python
# Create series containing only the cost for cleaning
cost = journals.iloc[:, 4]
cost
```

```python
# remove any value that is not a digit value
cost = cost.apply(lambda x: ''.join(list(filter(str.isdigit, str(x)))))
# make the cost series integers
cost = cost.apply(lambda x: int(x))
# return to decimal format
cost = cost/100
cost
```

```python
# return cleaned cost values into the data frame
journals.iloc[:, 4] = cost
journals
```

```python
# find cost mean
journals.groupby('Journal title')['COST (£) charged to Wellcome (inc VAT when charged)'].mean()
```

```python
# find cost median
journals.groupby('Journal title')['COST (£) charged to Wellcome (inc VAT when charged)'].median()
```

```python
#find cost standard deviation
journals.groupby('Journal title')['COST (£) charged to Wellcome (inc VAT when charged)'].std()
```
